# Introduction

In this analysis, I will explore the Iris dataset, a classic dataset in machine learning and statistics. It contains measurements of sepal length, sepal width, petal length, and petal width, along with the species of iris flowers (`setosa`, `versicolor`, and `virginica`). Using visualizations such as bar charts, pie charts, scatter plots, and histograms, we aim to uncover patterns, relationships, and distributions in the data, highlighting the distinctions between the flower species.

### Import the necessary libraries

```python
# Import libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import load_iris
```

### Load in the Iris Dataset from SKlearn inbuilt datasetss

```python
# Load the Iris dataset
iris = load_iris()
data = pd.DataFrame(iris.data, columns=iris.feature_names)
data['species'] = iris.target
data['species'] = data['species'].map({0: 'setosa', 1: 'versicolor', 2: 'virginica'})
```

```python
# Display the first few rows of the dataset
data.head()
```

### Exploratory Data Analysis (EDA)

```python
# Bar chart: Count of each species
plt.figure(figsize=(8, 5))
data['species'].value_counts().plot(kind='bar', color=['skyblue', 'lightgreen', 'salmon'])
plt.title('Class Distribution of Species')
plt.xlabel('Species')
plt.ylabel('Count')
plt.xticks(rotation=0)
plt.show()
```

```python
# Pie chart: Proportion of each species
plt.figure(figsize=(8, 5))
data['species'].value_counts().plot(kind='pie', autopct='%1.1f%%', colors=['skyblue', 'lightgreen', 'salmon'])
plt.title('Proportion of Each Species')
plt.ylabel('')
plt.show()
```

```python
# Scatter plot: Sepal Length vs Sepal Width
plt.figure(figsize=(8, 5))
sns.scatterplot(data=data, x='sepal length (cm)', y='sepal width (cm)', hue='species') 
plt.title('Sepal Length vs Sepal Width by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Sepal Width (cm)')
plt.legend(title='Species')
plt.show()
```

```python
# Histogram: Distribution of Sepal Length
plt.figure(figsize=(8, 5))
sns.histplot(data=data, x='sepal length (cm)', hue='species') # kde=True, hue='species') #palette='Set2', multiple="stack")
plt.title('Distribution of Sepal Length by Species')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Frequency')
plt.show()
```
